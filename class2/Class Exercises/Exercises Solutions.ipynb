```python
# Exercise 2: Find text/string with year "1952"

import requests
from bs4 import BeautifulSoup
import re

# Add headers to mimic a browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Fetch Wikipedia page with headers
url = "https://en.wikipedia.org/wiki/Data_processing"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

print("=== TEXT CONTAINING '1952' ===\n")

# Find all paragraphs and search for "1952" in them
paragraphs = soup.find_all('p')

count = 0
for paragraph in paragraphs:
    text = paragraph.get_text()
    if re.search(r'1952', text):
        count += 1
        print(f"{count}. {text.strip()}\n")

if count == 0:
    print("No results found with '1952'")
```

```python
# Exercise 3: Find text/string with year "1970" (limit to 2 results)

import requests
from bs4 import BeautifulSoup
import re

# Add headers to mimic a browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Fetch Wikipedia page with headers
url = "https://en.wikipedia.org/wiki/Database"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

print("=== TEXT CONTAINING '1970' (Limit: 2) ===\n")

# Find all paragraphs and headings (h1, h2, h3, h4, h5, h6)
elements = soup.find_all(['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6'])

count = 0
for element in elements:
    if count >= 2:  # Stop after finding 2 results
        break
    
    text = element.get_text()
    if re.search(r'1970', text):
        count += 1
        element_type = element.name.upper()
        print(f"{count}. [{element_type}] {text.strip()}\n")

if count == 0:
    print("No results found with '1970'")
```

```python
# Exercise 4: Fetch list of issues voted in US Senate

import requests
from bs4 import BeautifulSoup

# Fetch the webpage
url = "https://www.senate.gov/legislative/LIS/roll_call_lists/vote_menu_114_1.htm"
response = requests.get(url)
soup = BeautifulSoup(response.text, "lxml")

print("=== LIST OF ISSUES VOTED IN US SENATE ===\n")

# Find all tables and check their classes
tables = soup.find_all('table')
print(f"Found {len(tables)} tables\n")

# Find the correct table (try without class first)
table = soup.find('table')

if table:
    # Get all rows from the table
    rows = table.find_all('tr')
    
    count = 0
    # Loop through each row
    for row in rows[1:]:  # Skip the header row
        # Get all columns in the row
        columns = row.find_all('td')
        
        if len(columns) >= 3:  # If row has enough data
            count += 1
            vote_number = columns[0].text.strip()
            issue = columns[1].text.strip()
            result = columns[2].text.strip()
            
            print(f"{count}. Vote #{vote_number}")
            print(f"   Issue: {issue}")
            print(f"   Result: {result}\n")
else:
    print("No table found on the page")
```
