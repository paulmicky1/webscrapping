```python
import requests
from bs4 import BeautifulSoup

# Base URL
base_url = "https://books.toscrape.com/catalogue/page-{}.html"

book_titles = []

# Loop through all 50 pages
for page in range(1, 51):
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all book titles
    books = soup.find_all("h3")
    for book in books:
        title = book.find("a")["title"]
        book_titles.append(title)

print(f"Total Books Found: {len(book_titles)}\n")
for title in book_titles:
    print(title)
```

```python
# Import libraries
import requests
from bs4 import BeautifulSoup

# Step 1: Scrape all book titles from all pages
base_url = "https://books.toscrape.com/catalogue/page-{}.html"
titles = []

for page in range(1, 51):  # The site has 50 pages total
    url = base_url.format(page)
    response = requests.get(url)
    if response.status_code != 200:
        break
    soup = BeautifulSoup(response.text, "html.parser")
    books = soup.select("h3 a")
    for book in books:
        title = book.get("title")
        if title:
            titles.append(title)

# Step 2: Sort titles alphabetically
titles = sorted(titles)

# Step 3: Print all titles with numbers
for i, title in enumerate(titles, start=1):
    print(f"{i}. {title}")
```

```python

import requests
from bs4 import BeautifulSoup
from fpdf import FPDF
import time

# Step 1: Scrape all book titles from all pages
base_url = "https://books.toscrape.com/catalogue/page-{}.html"
titles = []

for page in range(1, 51):  # The site has 50 pages total
    url = base_url.format(page)
    try:
        response = requests.get(url)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching page {page}: {e}")
        continue
    
    soup = BeautifulSoup(response.text, "html.parser")
    books = soup.select("h3 a")
    for book in books:
        title = book.get("title")
        if title:
            titles.append(title)
    
    time.sleep(1)  # polite delay between requests

# Step 2: Sort titles alphabetically
titles = sorted(titles)

# Step 3: Save titles to a PDF
pdf = FPDF()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()
pdf.set_font("Arial", size=12)

for i, title in enumerate(titles, start=1):
    pdf.multi_cell(0, 10, f"{i}. {title}")

pdf_file_name = "book_titles.pdf"
pdf.output(pdf_file_name)

print(f"PDF created successfully: {pdf_file_name}")
```
